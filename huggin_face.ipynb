```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from huggingface_hub import notebook_login
from transformers import AutoTokenizer, AutoModelForSequenceClassification, AutoModelForTokenClassification, pipeline
```

```python
from transformers import pipeline
classifier = pipeline("zero-shot-classification",
                      model="facebook/bart-large-mnli")
```

```python
sent_1 = 'Virat Kohli is the best batsman in the world'
candidate_labels = ['Politics', 'Sports', 'Education', 'Art']
classifier(sent_1, candidate_labels)
```

```python
sent_1 = 'world war is a gruesome war'
candidate_labels = ['Politics', 'Sports', 'Education', 'Art']
classifier(sent_1, candidate_labels)
```

```python
from transformers import pipeline
classifier = pipeline("zero-shot-classification",
                      model="MoritzLaurer/deberta-v3-large-zeroshot-v2.0")
```

```python
sent_1 = 'We noticed a login to your account @ksanthosh23 from a new device. Was this you?'
candidate_labels = ['Ham', 'Spam']
classifier(sent_1, candidate_labels)
```

```python
model_name = "deepset/roberta-base-squad2"
```

```python
nlp = pipeline('question-answering', model=model_name, tokenizer=model_name)
QA_input = {
    'question': 'what are covalent bonds',
    'context': 'Its chemical formula, H2O, indicates that each of its molecules contains one oxygen and two hydrogen atoms, connected by covalent bonds'
}
res = nlp(QA_input)
```

```python
print(res)
```

```python
nlp = pipeline('question-answering', model=model_name, tokenizer=model_name)
QA_input = {
    'question': 'What is Jeppiar College',
    'context': 'It is a technical college offering courses in technology',
    'question': 'Where is Jeppiar College',
    'context': 'It is located in Chennai in OMR Road'}
res = nlp(QA_input)
```

```python
print(res)
```
